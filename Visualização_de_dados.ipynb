```python
#Gráfico de linhas

import matplotlib.pyplot as plt

x=[1,3,8]
y=[5,4,7]

titulo='Gráfico'

eixoX='Eixo X'
eixoY='Eixo Y'

plt.title(titulo)#
plt.xlabel(eixoX)#Legendas no gráfico
plt.ylabel(eixoY)#

plt.plot(x, y) #cria o gráfico de linhas
plt.show() #imprime o gráfico
```

```python
#gráfico de barras

x=[1,3,5,9]
y=[2,4,6,10]

titulo='Gráfico'

eixoX='Eixo X'
eixoY='Eixo Y'

plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.bar(x, y) #cria o gráfico de barras
plt.show() #imprime o gráfico
```

```python
#Comparação de gráficos de barra

x1=[1,3,5,7,9]
y1=[2,6,4,8,7]

x2=[2,4,6,8,10]
y2=[5,4,1,6,3]

titulo1='Gráfico de Comparação'
eixox='Eixo X'
eixoy='Eixo Y'

plt.title(titulo1)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1,y1,label='Grupo 1')
plt.bar(x2,y2, label='Grupo 2')
plt.legend()#mostra as legendas dos parametros "label"
plt.show()

```

```python
#Gráfico de dispersão

x=[1,3,5,7,9]
y=[2,6,4,8,7]

titulo='Gráfico de dispersão'
eixoX='Eixo X'
eixoY='Eixo Y'

plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.scatter(x,y, label='Pontos de disperção', color='black') #faz o gráfico de dispersão
plt.plot(x,y, color='red') #faz o gráfico de linhas, porém agora ligando os pontos
plt.show()
```
